```python
def createObjectRelation(name, domainType, rangeType):
    rel = URIRef(bhai[name])
    graph.add((rel, RDF.type, OWL.ObjectProperty))
    if domainType is not None:
        dom = bhai[domainType]
        graph.add((rel, RDFS.domain, dom))
    if rangeType is not None:
        ran = bhai[rangeType]
        graph.add((rel, RDFS.range, ran))
    return rel

def createClass(name, parentClass, labelFr, comment):
    concept = URIRef(bhai[name])
    graph.add((concept, RDF.type, OWL.Class))
    if parentClass is not None:
        graph.add((concept, RDFS.subClassOf, parentClass))
    if labelFr is not None:
        graph.add((concept, RDFS.label, Literal(labelFr, lang="fr")))
    if comment is not None:
        graph.add((concept, RDFS.comment, Literal(comment)))
    return concept

createClass("Personne", None, "Personne", None )
#createClass('Materiel', None, "Materiel", None )
createClass("Collection", None, "Collection", None )
#createClass("Titre", None, "Titre", None )
#createClass("Periode", None, "Periode", None )

createObjectRelation("appartient_a", "Sceau", "Personne")
createObjectRelation("a_materiel", "Sceau", 'Materiel')
createObjectRelation("a_collection", "Sceau", 'Collection')
createObjectRelation("a_titre", "Sceau", 'Titre')
createObjectRelation("est_periode", "Sceau", 'Periode')
```
